# Using pandas read_html to Web Scrape Data for Data Science

### Batting averages (current players)

```python
import pandas as pd
```

```python
url = 'https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=2;current=2;id=6;type=team'
url
```

```python
df = pd.read_html(url,header=0)
df
```

```python
len(df)
```

```python
players = df[0]
players
```

```python
players.to_csv("batting_avg.csv")
```
